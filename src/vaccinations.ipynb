```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
url = 'https://storage.googleapis.com/covid19-open-data/v3/vaccinations.csv'

vaccinations = pd.read_csv(url)
```

```python
vaccinations.head()
```

```python
vaccinations.info()
```

```python
vaccinations.describe()
```

```python
ax = sns.lineplot(data=vaccinations, x='date', y='cumulative_persons_vaccinated')
ax.set(xticklabels=[])
ax.set(xticks=[])
plt.show()
```
